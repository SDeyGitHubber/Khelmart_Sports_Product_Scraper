```python
import os
import urllib.request
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Define the base URL
base_url = 'https://www.khelmart.com/badminton/badminton-rackets?product_list_limit=60'

# Send a GET request to the base URL
response = requests.get(base_url)

# Parse the response content with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all img tags with class 'product-image-photo hovered-img'
img_tags = soup.find_all('img', {'class': 'product-image-photo main-img'})

# Initialize a list to store the image URLs
images = []

# Iterate over each img tag
for img_tag in img_tags:
    # Get the src attribute of the img tag
    img_url = img_tag.get('src')

    # Add the image URL to the images list
    images.append(img_url)

# Print the image URLs
for img_url in images:
    print(img_url)
```

```python
li_tags = soup.find_all('li', {'class': 'item product product-item'})

# Initialize a list to store the product names
product_names = []

# Iterate over each li tag
for li_tag in li_tags:
    # Find the strong tag with class 'product name product-item-name'
    strong_tag = li_tag.find('strong', {'class': 'product name product-item-name'})
    
    # If the strong tag is found
    if strong_tag is not None:
        # Find the a tag with class 'product-item-link'
        a_tag = strong_tag.find('a', {'class': 'product-item-link'})
        
        # If the a tag is found
        if a_tag is not None:
            # Get the text of the a tag and add it to the product names list
            product_names.append(a_tag.text.strip())

# Print the product names
for product_name in product_names:
    print(product_name)
    
print(f'\n\n {len(product_names)}')
```

```python
# Find all span tags with class 'old-price'
old_price_tags = soup.find_all('span', {'class': 'old-price'})

# Initialize a list to store the old prices
old_prices = []

# Iterate over each span tag
for span_tag in old_price_tags:
    # Find the span tag with class 'price'
    price_tag = span_tag.find('span', {'class': 'price'})
    
    # If the price tag is found
    if price_tag is not None:
        # Get the text of the price tag and add it to the old prices list
        old_prices.append(price_tag.text.strip())

# Find all span tags with class 'special-price'
special_price_tags = soup.find_all('span', {'class': 'special-price'})

# Initialize a list to store the special prices
special_prices = []

# Iterate over each span tag
for span_tag in special_price_tags:
    # Find the span tag with class 'price'
    price_tag = span_tag.find('span', {'class': 'price'})
    
    # If the price tag is found
    if price_tag is not None:
        # Get the text of the price tag and add it to the special prices list
        special_prices.append(price_tag.text.strip())

# Print the old prices and the special prices
print(f"Old Prices: {old_prices}")
print(f"Special Prices: {special_prices}")
```

```python
import pandas as pd

# Calculate the discount percentages
discounts = [(old - special) / old * 100 for old, special in zip(old_prices, special_prices)]

# Create a dataframe
df = pd.DataFrame({
    'Image': images,
    'Product Name': product_names,
    'Old Price': old_prices,
    'Special Price': special_prices,
    'Discount %': discounts
})

# Print the dataframe
print(df)
```
